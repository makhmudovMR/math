# functions and function's properties

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
print(matplotlib.__version__)
```

```python
'''линейная функция'''
x = np.arange(-15,15,0.1)
f = 2 * x + 1
plt.plot(x,f)
plt.grid(True)
plt.show()
```

```python
'''квадратичная функция'''
f = x**2
plt.plot(x,f)
plt.grid(True)
plt.show()
```

```python
'''кубическая функция'''
f = x**3
plt.plot(x,f)
plt.grid(True)
plt.show()
```

```python
'''функция'''
f = np.sqrt(x)
plt.plot(x,f)
plt.grid(True)
plt.show()
```
