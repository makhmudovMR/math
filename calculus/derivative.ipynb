```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

$$\lim Δx→0 \space \frac{f(x+Δx)-f(x)}{Δx}$$

```python
x_ = np.arange(-10, 11, 1)
f = x_**4

plt.plot(x_,f)
plt.grid(True)
plt.show()
```

```python
def f(x):
    return x**3

def derivative(x):
    delta_x = 1. / 1000.
    rise = f(x + delta_x) - f(x)
    slope = rise / delta_x
    return slope
```

```python
derivative(3)
```

```python
d = list(map(derivative, x_))
d
```

```python
for i in range(len(d)):
    plt.scatter(i, d[i])
plt.grid(True)
plt.show()
```
