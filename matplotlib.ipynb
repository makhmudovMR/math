# MATPLOTLIB

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(-10, 10, 20)
f_quadratic = x**2
```

```python
x
```

```python
plt.plot(x,f_quadratic)
plt.grid(True)
```

```python
f_cube = x**3
plt.plot(x,f_cube)
plt.grid()
```

```python
plt.plot(x, f_cube, color='blue', marker='o', linestyle='--', markerfacecolor='green');
plt.grid(True)
```

```python
plt.plot(x, f_cube, color='red', marker='o', linestyle='--', label='Линия 1')
plt.plot(x, f_quadratic, color='blue', marker='x', label='Линия 2')
plt.legend(loc='upper center', ncol=2)
plt.grid()

#Подписи для осей:
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

#Диапазон оси y:
plt.ylim(0, 120)

#Заголовок:
plt.title(r'График функций $y = x ^2 $ и $y = x^3$', fontsize=16, y=1.05);
```

```python
f_sigm = 1/(1+np.exp(-x))
plt.plot(x,f_sigm)
plt.grid()
```

```python
f_error = (x*2-x)**2
plt.plot(x, f_error)
```
